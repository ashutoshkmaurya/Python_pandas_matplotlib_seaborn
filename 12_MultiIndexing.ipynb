Plotting with Unstack()

```python
df
```

```python
df.groupby(["pclass","sex"]).mean()
```

```python
df.groupby(["pclass","sex"])["age"].mean()
```

```python
df.groupby(["pclass","sex"])["age"].mean().plot(kind="bar")
```

```python
df.groupby(["pclass","sex"])["age"].mean().unstack()
```

```python
df.groupby(["pclass","sex"])["age"].mean().unstack().plot(kind="bar")
```

```python
df.groupby(["pclass","sex"])["age"].mean()
```

```python
df.groupby(["pclass","sex"])["age"].mean().unstack(level="pclass").plot(kind="bar")
```

```python
df.groupby(["sex","survived"]).mean()
```

```python
df.groupby(["sex","survived"])["age"].mean().plot(kind="bar")
```

```python
df.groupby(["sex","survived"])["age"].mean().unstack()
```

```python
df.groupby(["sex","survived"])["age"].mean().unstack().plot(kind="bar")
```

```python
df.groupby(["sex","survived"])["age"].mean().unstack().rename(columns={0:"Died",1:"Survived"})
```

```python
df.groupby(["sex","survived"])["age"].mean().unstack().rename(columns={0:"Died",1:"Survived"}).plot(kind="bar")
```

```python
def get_age_group(age):
    if age < 2 :
        return "infant"
    elif age < 12:
        return "child"
    elif age < 18:
        return "teen"
    elif age < 50:
        return "adult"
    else:
        return "senior"
```

```python
get_age_group(1)
```

```python
get_age_group(10)
```

```python
get_age_group(17)
```

```python
titanic["age"].apply(get_age_group)
```

```python
titanic["age_group"] = titanic["age"].apply(get_age_group)
```

```python
titanic
```

```python
titanic.age_group.value_counts()
```

```python
titanic.groupby("age_group").survived.mean()
```

```python
titanic.groupby(["sex", "age_group"]).survived.mean()
```

```python
titanic.groupby(["age_group","sex"]).survived.mean()
```

```python
def get_fam_size(s):
    fam_size = s.sibsp + s.parch
    if fam_size == 0:
        return "solo"
    elif fam_size < 5:
        return "average"
    else:
        return "large"
```

titanic.apply(get_fam_size, axis=1)

```python
titanic["fam_size"] = titanic.apply(get_fam_size, axis=1)
```

```python
titanic["fam_size"].value_counts()
```

```python
titanic.groupby("fam_size").survived.mean()
```

```python
titanic.groupby(["fam_size","sex"]).survived.mean()
```

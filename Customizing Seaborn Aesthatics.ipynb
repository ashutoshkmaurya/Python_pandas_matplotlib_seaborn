# Altering Spines With despine()


```python
sns.set_style("white")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.histplot(data=tips, x="total_bill")
```

```python
sns.histplot(data=tips, x="total_bill")
sns.despine()
```

```python
sns.histplot(data=tips, x="total_bill")
```

```python
sns.histplot(data=tips, x="total_bill")
sns.despine(bottom=True)
```

```python
sns.histplot(data=tips, x="total_bill")
sns.despine(bottom=True, top=False)
```

Using The Aggregate Method

pandas.core.groupby.DataFrameGroupBy.aggregate


```python
df.groupby("sex").agg("min")
```

```python
titanic.groupby("sex")["age"].agg("min")
```

```python
titanic.groupby("sex")["age"].agg(["min","max","mean", "median"])
```

```python
titanic.groupby("sex").agg({"age":["min","max"],"pclass":"mean"})
```

```python
df.groupby("sex").agg("mean")
```

```python
carstocks
```

```python
carstocks.groupby("Symbol").agg({"Open":"mean","Close":"mean"})
```

```python
carstocks.groupby("Symbol").agg({"Open":"mean","Close":"mean","Volume":["mean","sum"]})
```

```python
titanic.groupby("pclass")["age"].agg(["min","max"])
```

```python
def range(x):
    return x.max() - x.min()

titanic.groupby("pclass")["age"].agg(range)
```

```python
def range(x):
    return x.max() - x.min()

titanic.groupby("pclass")["age"].agg(["min","max",range])
```

```python
titanic["age"].size - titanic["age"].count()
```

```python
def count_nulls(s):
    return s.size - s.count()

titanic.groupby("pclass").agg(count_nulls)
```

```python
def count_nulls(s):
    return s.size - s.count()

titanic.groupby("pclass")["age"].agg(count_nulls)
```
